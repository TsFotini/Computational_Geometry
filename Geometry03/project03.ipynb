#### Exercise 1
Compute Voronoi diagrams of diﬀerent sets of vertices of your choice using the routine Voronoi (and its companion voronoi plot 2d for visualization) from the module scipy.spatial. Plot your results.

In order to understand the code , we need to introduce the definition of a Voronoi diagram. A Voronoi diagram is a partition of a plane into regions close to each of a given set of objects.For each seed there is a corresponding region consisting of all points of the plane closer to that seed than to any other. These regions are called Voronoi cells. The Voronoi diagram of a set of points is dual to its Delaunay triangulation.
The pointlist variable, is actually the list of coordinates of points to construct a convex hull from. Below we present another example of a new list of points.

```python
import numpy as np 
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
```

```python
#(0, 3) (0, 0) (3, 0) (3, 3)
pointlist = np.array([[0,3], [0,0], [3,0], [3,3]])
vor = Voronoi(pointlist)
voronoi_plot_2d(vor)
plt.show()
```

```python
import matplotlib as mpl
import matplotlib.cm as cm

colors_dictionary = ['blue','green','red','gray','magenta','yellow','purple','burlywood','cyan','pink','brown']
colors = []
voronoi_plot_2d(vor2, show_points=True, show_vertices=False, s=1)
for r in range(len(pr)):
    if len(colors) == 0:
        colors = colors_dictionary.copy()
    region = vor2.regions[pr[r]]
    if not -1 in region:
        polygon = [vor2.vertices[i] for i in region]
        plt.fill(*zip(*polygon), color=colors[0])
        colors.pop(0)
plt.show()
```

#### Exercise 2
Using the routine Delaunay in the module scipy.spatial compute the Delaunay triangulation of diﬀerent sets of vertices of your choice and plot your results.

The first example of data we give, is a set of points that are colinear. For a set of points on the same line there is NO Delaunay triangulation as we can see below. 

```python
from scipy.spatial import Delaunay
straight_line_points = np.array([[0,0], [1,1], [2,2], [3,3]])
plt.plot(straight_line_points[:,0], straight_line_points[:,1], 'o')
plt.show()
```

So the next step, is to insert a set of points that are not colinear. Furthermore this particular set is a set of points that constructs a convex hull.

```python
plist = np.array([[0,3], [0,0], [3,0], [3,3]])
tri = Delaunay(plist)
plt.triplot(plist[:,0], plist[:,1], tri.simplices)
plt.plot(plist[:,0], plist[:,1], 'o')
plt.show()
```

If we desire to add one more data point to our initial set the results will be those presented below.

```python
def delaunay_(plist):
    tri = Delaunay(plist)
    plt.triplot(plist[:,0], plist[:,1], tri.simplices)
    plt.plot(plist[:,0], plist[:,1], 'o')
    plt.show() 
    return tri
new_element = np.array([8,8])
new_plist = np.vstack((plist, new_element))
new_plist
```

```python
delaunay_(new_plist)
```

```python
from scipy.sparse.csgraph import shortest_path
Distances, Pred = shortest_path(newarray, directed=False, method='FW', return_predecessors=True)
```

```python
def get_path(Pr, i, j,pointlist):
    pointspath = []
    path = [j]
    k = j
    while Pr[i, k] != -9999:
        path.append(Pr[i, k])
        k = Pr[i, k]
    path = path[::-1]
    for p in path:
        pointspath.append(pointlist[p])
    return pointspath
get_path(Pred,0,1,pointlist)
```
